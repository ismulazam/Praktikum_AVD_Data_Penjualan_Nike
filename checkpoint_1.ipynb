#**BUSINESS UNDERSTANDING**

## **Study Case**

Nike adalah salah satu perusahaan sepatu terbesar di dunia dengan penjualan yang sangat tinggi di berbagai negara. Namun, meskipun penjualannya besar, perusahaan tetap menghadapi masalah seperti keuntungan yang tidak selalu stabil, adanya produk yang kurang menguntungkan, serta pemberian diskon yang belum tentu meningkatkan laba. Jika hal ini tidak dianalisis dengan baik, perusahaan bisa mengalami penurunan keuntungan meskipun penjualan terlihat ramai. Oleh karena itu, data penjualan perlu dianalisis untuk mengetahui produk, wilayah, dan cara penjualan mana yang paling menguntungkan agar Nike tetap konsisten dan terus mendapatkan keuntungan.

## **Business Objective**

Tujuan dari analisis ini adalah membantu Nike menjaga keuntungan tetap stabil meskipun persaingan tinggi dan strategi diskon sering digunakan. Perusahaan ingin mengetahui produk mana yang paling menguntungkan, wilayah mana yang memberikan kontribusi terbesar, serta saluran penjualan (Online atau Retail) mana yang lebih efektif. Selain itu, Nike ingin memastikan bahwa strategi diskon yang diberikan benar-benar meningkatkan penjualan tanpa mengurangi laba secara signifikan.

Dataset ini berisi data penjualan produk retail yang mencakup informasi seperti produk, wilayah, saluran penjualan (Online dan Retail), jumlah unit terjual, diskon, revenue, dan profit. Data ini digunakan untuk melihat kinerja penjualan serta menganalisis keuntungan perusahaan berdasarkan berbagai faktor yang memengaruhinya.


## **Assess Situation**

Saat ini, Nike memiliki banyak data penjualan yang mencatat produk yang terjual, wilayah penjualan, cara penjualan (Online atau Retail), jumlah barang yang terjual, diskon yang diberikan, serta keuntungan yang didapat. Namun dari data tersebut terlihat bahwa keuntungan tidak selalu stabil. Ada beberapa penjualan yang keuntungannya kecil bahkan ada yang rugi. Selain itu, belum diketahui dengan jelas apakah diskon yang diberikan benar-benar membantu meningkatkan keuntungan atau justru mengurangi laba. Jika kondisi ini dibiarkan saja, perusahaan bisa terus mengambil keputusan yang kurang tepat. Namun, karena data penjualan sudah tersedia, perusahaan bisa menganalisis dan memiliki peluang besar untuk memanfaatkannya agar strategi penjualan menjadi lebih efektif dan menguntungkan.
