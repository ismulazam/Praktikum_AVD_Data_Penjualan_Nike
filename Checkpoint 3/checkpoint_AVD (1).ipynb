### **Outliers Values**

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col} → {len(outliers)} outliers")
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Temuan:
* Revenue max = 37169,35 (sangat tinggi dibanding mean 274,87)
* Std Revenue (2023,77) jauh lebih besar dari mean
* Profit max = 3999,21
* Units_Sold max = 4 (relatif kecil)

Insight:

Revenue sangat right-skewed. Ada transaksi bernilai sangat besar yang menarik rata-rata naik drastis.
Ini menunjukkan kemungkinan:
* Transaksi korporat
* Penjualan bundle besar
* Atau anomali input

Kesimpulan:

Outlier pada Revenue dan Profit perlu dianalisis:
* Jika valid → itu high-value customer (peluang bisnis besar)
* Jika tidak valid → perlu dibersihkan agar model tidak bias

*   **Bukan Hubungan Kuat:** Korelasi ini moderat, bukan sangat kuat (-1). Ini menunjukkan bahwa diskon adalah salah satu faktor yang mempengaruhi keuntungan, tetapi bukan satu-satunya. Ada faktor lain yang juga berperan, atau hubungan tersebut tidak sepenuhnya linear.

Interpretasi Bisnis:

*   **Evaluasi Strategi Diskon:** Nike perlu mengevaluasi kembali strategi diskonnya. Diskon mungkin menarik pelanggan dan meningkatkan revenue (pendapatan kotor), tetapi secara signifikan memakan profit (keuntungan bersih).
*   **Optimalisasi Titik Diskon:** Perusahaan harus mencari 'titik manis' di mana diskon masih efektif menarik pelanggan tanpa terlalu banyak menggerus keuntungan. Ini mungkin melibatkan A/B testing berbagai tingkat diskon.
*   **Fokus pada Nilai:** Selain diskon, Nike bisa mempertimbangkan untuk menekankan nilai produk, kualitas, atau fitur unik untuk menarik pelanggan, sehingga tidak terlalu bergantung pada harga rendah.
*   **Segmentasi Pelanggan:** Mungkin ada segmen pelanggan yang sensitif terhadap harga dan merespons diskon, sementara ada segmen lain yang lebih menghargai brand atau kualitas. Strategi diskon bisa lebih ditargetkan.

####**Outliers Values**

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col} → {len(outliers)} outliers")
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan hasil pengecekan outliers, kolom Units_Sold memiliki outliers 16.68%, MRP 49.84%, dan Discount_Applied 12.56%. Kolom Revenue berhasil ditangani dari yang sebelumnya memiliki outliers menjadi 0%.
